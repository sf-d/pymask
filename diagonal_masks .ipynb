```python
def polygon_intersect (polygons):
    all_list = []
    for p in polygons:
        poly_list =[]
        for pp in polygons:
            if p.covers(pp) and polygons.index(p)!=polygons.index(pp):
                if p in poly_list:
                    poly_list.append(pp)
                else:
                    poly_list = list(chain([p,pp]))
            else:
                pass
        all_list.append(poly_list)
    all_list = [x for x in all_list if x!=[]]
    for i in polygons:
        if any(i in sl for sl in all_list):
            pass
        else:
            all_list.append([i])
    return all_list
```
